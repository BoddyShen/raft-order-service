```python
import matplotlib.pyplot as plt
```

# Latencies With and Without Cache

```python
# Data representing the probability to send order request after query request and their corresponding average latencies
probability = ["0%", "20%", "40%", "60%", "80%"]
query_latencies_with_cache = [0.15255698, 0.15876184, 0.16180804, 0.16341730, 0.16486630]
buy_latencies_with_cache = [None, 0.26665375, 0.24241932, 0.24599216, 0.24065582]
query_latencies_no_cache = [0.15740024, 0.16054750, 0.16275914, 0.17821876, 0.17012994]
buy_latencies_no_cache = [None, 0.20297315, 0.20367602, 0.23548602, 0.21326166]


# Creating the plot
plt.figure(figsize=(10, 6))  # Set the figure size for better readability
plt.plot(probability, query_latencies_with_cache, label='Query (With Cache)', marker='o', linestyle='-', color='blue')  # Plot Query latencies with cache
plt.plot(probability, buy_latencies_with_cache, label='Buy (With Cache)', marker='v', linestyle='-', color='blue')  # Plot Buy latencies with cache
plt.plot(probability, query_latencies_no_cache, label='Query (Without Cache)', marker='o', linestyle='-', color='red')  # Plot Query latencies without cache
plt.plot(probability, buy_latencies_no_cache, label='Buy (Without Cache)', marker='v', linestyle='-', color='red')  # Plot Buy latencies without cache

# Adding labels and title to the plot
plt.xlabel('Order Probability')  # X-axis label
plt.ylabel('Average Latency (seconds)')  # Y-axis label
plt.title('Average Latencies for Query and Buy Actions')  # Plot title

# Adding a legend to explain the markers
plt.legend()

# Adding a grid for better readability of the plot
plt.grid(True)

# Ensuring all client number ticks are shown
plt.xticks(probability)

# Display the plot
plt.show()
```
